# topic modeling

For this exercise, we will develop a topic model that allows us to examine if there are any significant differences between the kind of topics mentioned for property descriptions that allude to a local park and those that do not (park_tf True vs. False). This may give us some insight on how landlords or agents market certain types of properties and if the adjacency of a park has any effect.

```python
import pandas as pd
listings = pd.read_pickle('combined_df_sorted.pkl')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

# set stopwords variable
swords = stopwords.words('english')

# define list of words that does not include words found in the stopwords list and is True for park_tf
wordlists_true = [[word for word in word_tokenize(desc) 
              if word not in swords] 
              for desc, park_tf in zip(listings['description'], listings['park_TF']) if park_tf == True]

# create single list instead of list of lists
wordlists_true = [word for sublist in wordlists_true for word in sublist]

# define list of words that does not include words found in the stopwords list and is False for park_tf
wordlists_false = [[word for word in word_tokenize(desc) 
              if word not in swords] 
              for desc, park_tf in zip(listings['description'], listings['park_TF']) if park_tf == False]

# create single list instead of list of lists
wordlists_false = [word for sublist in wordlists_false for word in sublist]
```

### Bag of Words

Let's first conduct some simple analysis that examines the prevalence of words for each list of words.

```python
counts_true = {}
# for loop to count each word and store its total
for word in wordlists_true:
    if word in counts_true:
        counts_true[word] +=1
    else:
        # doesn't exist in the dictionary
        counts_true[word] = 1

counts_false = {}
for word in wordlists_false:
    if word in counts_false:
        counts_false[word] +=1
    else:
        # doesn't exist in the dictionary
        counts_false[word] = 1
```

```python
# store true word count as dataframe
df_true = pd.DataFrame.from_dict(counts_true, orient='index', columns=['word_count'])
df_true.sort_values('word_count', ascending=False, inplace=True)
df_true.index.name = 'word'
# store false word count as dataframe
df_false = pd.DataFrame.from_dict(counts_false, orient='index', columns=['word_count'])
df_false.sort_values('word_count', ascending=False, inplace=True)
df_false.index.name = 'word'
```

```python
# check dataframe
df_true.head(20)
```

There are high-frequency words that do not appear to add much value to our analysis and therefore we will drop them. 

```python
# drop top words that are generic to the area or nature of property descriptions
df_true.drop(["los",
              "angeles", 
              "apartments",
              "apartment",
              "la",
              "contact",
              "info",
              "call",
              "show", 
              "housing",
              "parking",
              "ca",
              "us",
              "https",
              "site",
              "availability",
              "available"], inplace = True)

df_false.drop(["los",
              "angeles", 
              "apartments",
              "apartment",
              "la",
              "contact",
              "info",
              "call",
              "show", 
              "housing",
              "parking",
              "ca",
              "us",
              "https",
              "site",
              "availability",
              "available"], inplace = True)
```

So although we see which words are most prevalent, it is not as helpful as it could be. It would be helpful to compare the most frequent words of the true category and compare them to the prevalency of the same words in the false category. This way we can better understand how they depart from each other. We'll do this using a stacked bar chart but we'll first convert the count of each word into a ratio of its count against the total word count. This will allow us to compare % differences which is necessary due to the fact that the sizes of each category differ greatly.

```python
# convert to percent of total word counts
df_true['percent'] = round((df_true['word_count']/df_true['word_count'].sum())*100,2)
df_false['percent'] = round((df_false['word_count']/df_false['word_count'].sum())*100,2)

df_true['percent_diff'] = round((df_true['word_count']/df_true['word_count'].sum())*100,2)
df_false['percent_diff'] = round((df_false['word_count']/df_false['word_count'].sum())*100,2)
```

```python
# join dataframes back together
df_true_false = df_true.join(df_false, how = "outer", lsuffix = "_true",rsuffix = "_false")
# sort and only keep top 20 of true category
df_true_false_limited_true = df_true_false.sort_values("percent_true", ascending = False).head(20)
df_true_false_limited_false = df_true_false.sort_values("percent_false", ascending = False).head(20)
```

```python
fig, ax = plt.subplots(2,1,figsize=(12, 8))

# Plot the double bar chart
sns.barplot(ax=ax[0], data=df_true_false_limited_true, x=df_true_false_limited_true.index, y="percent_true", color = "#ADD8E6", label="True")
sns.barplot(ax=ax[0], data=df_true_false_limited_true, x=df_true_false_limited_true.index, y="percent_false", color="#D8BFD8", label="False", bottom=df_true_false_limited_true["percent_true"])

# Set x-tick labels
#ax.set_xticks(df_true_false_limited.index)
ax[0].set_xticklabels(df_true_false_limited_true.index)

# Set labels and title
ax[0].set_xlabel('words')
ax[0].set_ylabel('% of total word count')
ax[0].set_title('')

# Rotate x-axis tick marks 
ax[0].tick_params(axis='x', rotation=30, labelsize = 8)
ax[0].tick_params(axis='y', labelsize = 8)
ax[0].legend()

# do the same but with false category
# Plot the double bar chart
sns.barplot(ax=ax[1], data=df_true_false_limited_false, x=df_true_false_limited_false.index, y="percent_true", color = "#ADD8E6", label="True", bottom=df_true_false_limited_false["percent_false"])
sns.barplot(ax=ax[1], data=df_true_false_limited_false, x=df_true_false_limited_false.index, y="percent_false", color="#D8BFD8", label="False")

# Set x-tick labels
#ax.set_xticks(df_true_false_limited.index)
ax[1].set_xticklabels(df_true_false_limited_false.index)

# Set labels and title
ax[1].set_xlabel('words')
ax[1].set_ylabel('% of total word count')
ax[1].set_title('')

# Rotate x-axis tick marks 
ax[1].tick_params(axis='x', rotation=30, labelsize = 8)
ax[1].tick_params(axis='y', labelsize = 8)

ax[1].legend()
```

### Topic Modeling

We will further explore the discrepancies in property descriptions by building out a topic model to understand how the two sets differ. 

```python
# remove words from original lists
remove_words = ["los",
                "angeles", 
                "apartments",
                "apartment",
                "la",
                "contact",
                "info",
                "call",
                "show", 
                "housing",
                "parking",
                "ca",
                "us",
                "https",
                "site",
                "availability",
                "available"]

wordlists_true = [word for word in wordlists_true if word not in remove_words]
wordlists_false = [word for word in wordlists_false if word not in remove_words]
```

```python
import gensim
# tokenize and run topic model
wordlists_tokenized_true = [word_tokenize(text) for text in wordlists_true]
dictionary_true = gensim.corpora.Dictionary(wordlists_tokenized_true)
corpus_true = [dictionary_true.doc2bow(word) for word in wordlists_tokenized_true]
model_true = gensim.models.LdaModel(corpus_true, id2word=dictionary_true, num_topics=4)

# tokenize again and rerun for wordlists_false list
wordlists_tokenized_false = [word_tokenize(text) for text in wordlists_false]
dictionary_false = gensim.corpora.Dictionary(wordlists_tokenized_false)
corpus_false = [dictionary_false.doc2bow(word) for word in wordlists_tokenized_false]
model_false = gensim.models.LdaModel(corpus_false, id2word=dictionary_false, num_topics=4)
```

#### Topic modeling for property descriptions that allude to local parks

```python
# generate topics for true list
for topic_id, topic_words in model_true.print_topics():
    print(f"Topic ID: {topic_id}\nWords: {topic_words}\n")
```

```python
# check first topic key words
model_true.show_topic(0)
```

```python
import pyLDAvis
import pyLDAvis.gensim_models

# visualize topics for wordslist_true
pyLDAvis.enable_notebook()
pyLDAvis.gensim_models.prepare(model_true, corpus_true, dictionary_true)
```

#### Topic modeling for property descriptions that do not allude to local parks

```python
# generate topics for false list
for topic_id, topic_words in model_false.print_topics():
    print(f"Topic ID: {topic_id}\nWords: {topic_words}\n")
```

```python
# check first topic key words
model_false.show_topic(0)
```

```python
# visualize topics for wordslist_false
pyLDAvis.enable_notebook()
pyLDAvis.gensim_models.prepare(model_false, corpus_false, dictionary_false)
```

#### Observations
- We've successfully produced a bag of words and conducted topic modeling for the property descriptions included in listing urls.
- Topics are equally distributed in size and have similar separation among the principal components, demonstrating that not one topic was more prevalent than others or that our topics had much overlap internally.
- Even when tinkering with the number of topics the model produced, it was difficult to produce clearly defined and segmented topics. 
- Both lists have multiple topics that touch upon amenities in and out unit, however, it appears from the eye-test that the property descriptions that mention local parks have a higher likelihood to focus on amenities out of unit/local to the area while the property descriptions that do not mention local parks have a higher likelihood to focus on amenities in unit/specific the apartment complex. 
- There is some distinction between the groups in using language that elicits more natural elements and accessibilty.
- It is difficult to glean too much from our topic modeling, unfortunately. Although we took steps to remove unhelpful words, there are still more words that remain that seem to clutter our analysis. We could continue to remove filler words from our analysis but this also comes at a cost of continuously limiting the words being analyzed. It would appear that this is a significant downfall to utilizing property descriptions for topic modeling as they really only contain cookie-cutter language. There is a lot of overlap between the most frequent words and topics covered, making it difficult to definitively separate out differences between the two groups we are analyzing.
